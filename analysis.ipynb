# Análise dos Resultados da Extração de Conflitos

```python
import pandas as pd
import pyarrow.parquet as pq
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar pd options para melhor visualização
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

## 1. Carregando as Tabelas Extraídas

```python
data_dir = "data/nature_of_agent_conflicts"

df_im = pd.read_parquet(f"{data_dir}/resolver_labels.parquet")
try:
    df_cc = pd.read_parquet(f"{data_dir}/conflict_chunks.parquet")
    df_rc = pd.read_parquet(f"{data_dir}/resolved_chunks.parquet")
    df_class = pd.read_parquet(f"{data_dir}/classified_chunks.parquet")
    has_chunks = True
except FileNotFoundError:
    has_chunks = False
    print("Tabelas de chunks não foram encontradas (Nenhum chunk de conflito foi gerado ainda).")
```

## 2. Inspeção de Internal Merges e Resolvers

```python
print(f"Total de merge commits internos: {len(df_im)}")

# Exibir a distribuição de resolução (Humanto / Bot)
display(df_im['resolver_type'].value_counts())

df_im.head()
```

## 3. Inspeção das Estratégias de Resolução (Stage 5)

```python
if has_chunks:
    print(f"Total de chunks de conflitos analisados: {len(df_class)}")
    
    plt.figure(figsize=(10, 5))
    sns.countplot(y='strategy', data=df_class, order=df_class['strategy'].value_counts().index)
    plt.title('Distribuição das Estratégias de Resolução')
    plt.show()
    
    display(df_class.head())
```
